# 1. Distribution of negatively charged particles
## (a) q-nonextensive distribution of electrons
The nonextensive entropy proposed by Tsallis is defiend as
\begin{equation}
S_{q} = k_{B}\frac{1-\sum_{k=1}^{N}\left(p_{k}\right)^{q}}{q-1} ...................(1)
\end{equation}

Nonextensive statistical Tsallis velocity distribution function for electrons is given by
\begin{equation}
f^\mathrm{e}\left(x,\textbf{v}^e\right) = A^e \left[1-\left(q-1\right) \left\lbrace \frac{v_{x}^2+v_{y}^2+v_{z}^2}{v_{t}^2}-\frac{e\phi(x)}{T^e}\right\rbrace \right]^\frac{1}{q-1}...................(2)
\end{equation}

As the parameter $q \rightarrow 1$, Tsallis entropy defined by equation (1) becomes
\begin{equation}
S_q = -k_B \sum_{k = 1}^N p_k \mathrm{ln}p_k  .......................(3)
\end{equation}

The Maxwellian velocity distribution fucntion is
\begin{equation}
f^{\mathrm{e}} \left( x, \textbf{v}^{\mathrm{e}} \right) = A^{\mathrm{e}} \exp\left[-\frac{v_{x}^2+v_{y}^2+v_{z}^2}{v_{t}^2}+\frac{e\phi(x)}{ T^{\mathrm{e}}}\right] ....................(4)
\end{equation}

## (b) Maxwellian distribution of negative ions 

The Maxwellian distribution function for negative ions is
\begin{equation}
f^{\mathrm{n}} \left( x, \textbf{v}^{\mathrm{n}} \right) = A^{\mathrm{n}} \exp\left[-\frac{v_{x}^2+v_{y}^2+v_{z}^2}{v_{t}^2}+\frac{e\phi(x)}{ T^{\mathrm{n}}}\right]  ......................(5)
\end{equation}

Spatial density distribution of negatively charged particle are
\begin{equation}
n\left(\phi\right)= \int\int \int f\left(x,\textbf{v}\right) d^3\textit{v} ...................(6)
\end{equation}
which gives as;
\begin{equation}
n^{e}\left(\phi\right) = n_{e0} \left(1+(q-1) \frac{e\phi(x)}{T^e}\right)^\frac{3q-1}{2q-2} ................(7)
\end{equation}

and

\begin{equation}
n^{n}\left(\phi\right) = n_{n0} \exp\left(\frac{e\phi(x)}{T^{n}} \right)  .......................(8)
\end{equation}

Define the transfromation equations as;

\begin{equation}
 \eta = -\frac{e\phi}{T^{e}},\hspace{0.3cm} N^{e} = \frac{n^{e}}{n_{e0}},\hspace{0.3cm} N^{n} = \frac{n^{n}}{n_{e0}},\hspace{0.3cm} \beta = \frac{T^{e}}{T^{n}}\hspace{0.3cm} \textrm{and} \hspace{0.3cm} \delta_{n} = \frac{n_{n0}}{n_{e0}} ................(9)
\end{equation}
 
 With these transformation relation, equations (8) and (9) becomes;
 \begin{equation}
N^{e}\left(\phi\right) = \left(1-(q-1)\eta\right)^\frac{3q-1}{2q-2} ................(10)
\end{equation}

and

\begin{equation}
N^{n}\left(\phi\right) = \delta_{n}\exp\left(-\eta\beta\right)  .......................(11)
\end{equation}
